```python
df_country = pd.read_excel('data/Country-Code.xlsx')
df_country.head()
```

```python
# combine with df having country code column
df.columns
```

```python
final_df = pd.merge(df,df_country,on='Country Code', how='left')
```

```python
final_df.head()
```

```python
#check data types
final_df.dtypes
```

```python
final_df.columns
```

```python
country_names = final_df.Country.value_counts().index
```

```python
country_values = final_df.Country.value_counts().values
```

```python
# pie chart - Top 3 countries that uses zomato
plt.pie(x=country_values[:3], labels=country_names[:3],autopct="%1.2f%%")
```

Observation: Zomato maximum records or transactions are from India around 94.39%.
    After that USA and United Kingdom 4.75% and 0.87% respectively.

```python
ratings = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating count'})
```

```python
ratings
```

## Observation
1. When rating is between 4.5 to 4.9 ---> Excellent
2. When rating is between 4.0 to 4.5 ---> Very good
3. When rating is between 3.5 to 3.9 ---> Good
4. When rating is between 2.5 to 3.4 ---> Average
5. When rating is between 1.8 to 2.4 ---> Poor

```python
# Bar plot
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6)
sns.barplot(x='Aggregate rating', y='Rating count', data=ratings)
```

```python
sns.barplot(x='Aggregate rating', y='Rating count',hue='Rating color', data=ratings, palette=['blue','red','orange','yellow','green','green'])
```

Observation:
1. Not rated count is very high
2. Maximum number of ratings are between 2.5 to 3.4

```python
## Count plot
sns.countplot(x='Rating color', data=ratings,palette=['blue','red','orange','yellow','green','green'])
```

```python
### find the countries name that has given 0 rating
final_df[final_df['Aggregate rating']==0].groupby('Country').size().reset_index()
# final_df[final_df['Rating color']=='White'].groupby('Country').size().reset_index()
# final_df[final_df['Aggregate rating']==0]['Country'].value_counts()
```

Observations: Maximum umber of 0 ratings are from Indian Customers

```python
# Find out which currency is used by which country
final_df[['Country','Currency']].groupby(['Country','Currency']).size().reset_index()
```

```python
# which country do have online deliveries option
final_df[['Country','Has Online delivery']].groupby(['Country','Has Online delivery']).size().reset_index()
```
